## Importing Libraries

```python
'''import supporting libraries'''

import numpy as np
import matplotlib.pyplot as plt
import imageio

import sys
sys.path.append('./code/') # add path to source code directory
```

## Preparing Data

```python
'''preparing the data for 2 sample crops'''

### For Rice
cost_rice = np.array([28000.0, 28500,29000, 26000]) # THB/ha
yield_rice = np.array([2.1, 2.35, 2.82, 3.22]) # ton/ha
farm_price_rice = np.arange(8.5,12,0.5)*1000 #THB/ton
yield_map_rice = imageio.imread('./sample_data/mod6_data/Rice_CropSuitability_rain.tif')/1000 ## in Ton


### For Maize
cost_maize = np.array([30100, 30000,27000,25500])
yield_maize = np.array([4.8,5.5,6.1,8.7])/ 2
farm_price_maize = np.arange(6.5,9.5,0.5)*1000 #THB/ton
yield_map_maize = imageio.imread('./sample_data/mod6_data/Maize_CropSuitability_rain.tif')/1000 ## in Ton
```

## Economic Suitability Analysis

```python
'''importing library and passing data'''

import EconomicSuitability
econ_su = EconomicSuitability.EconomicSuitability()

econ_su.addACrop(crop_name='rice', crop_cost=cost_rice, crop_yield=yield_rice, farm_price=farm_price_rice, yield_map=yield_map_rice)
econ_su.addACrop(crop_name='maize', crop_cost=cost_maize, crop_yield=yield_maize, farm_price=farm_price_maize, yield_map=yield_map_maize)
```

```python
'''get result'''

rice_rev = econ_su.getNetRevenue('rice') # THB/ha
rice_rev_class = econ_su.getClassifiedNetRevenue('rice')
rice_rev_norm = econ_su.getNormalizedNetRevenue('rice')

maize_rev = econ_su.getNetRevenue('maize') # THB/ha
maize_rev_class = econ_su.getClassifiedNetRevenue('maize')
maize_rev_norm = econ_su.getNormalizedNetRevenue('maize')
```

```python
'''visualize result'''

print('For Rice')

plt.imshow(rice_rev)
plt.colorbar()
plt.show()
plt.imshow(rice_rev_class)
plt.colorbar()
plt.show()
plt.imshow(rice_rev_norm)
plt.colorbar()
plt.show()

print('For Maize')

plt.imshow(maize_rev)
plt.colorbar()
plt.show()
plt.imshow(maize_rev_class)
plt.colorbar()
plt.show()
plt.imshow(maize_rev_norm)
plt.colorbar()
plt.show()
```
